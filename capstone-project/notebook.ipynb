```python
import torch
import torch.nn as nn
from torchvision import models

# 1. –ù–∞—Å—Ç—Ä–æ–π–∫–∏
IMG_SIZE = 224
MODEL_PATH = "leukemia_efficientnet_optimized.pth"
ONNX_PATH = "leukemia_model.onnx"

# 2. –°–æ–∑–¥–∞–µ–º –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä—É –º–æ–¥–µ–ª–∏ (—Ç–∞–∫—É—é –∂–µ, –∫–∞–∫ –ø—Ä–∏ –æ–±—É—á–µ–Ω–∏–∏)
model = models.efficientnet_b0(weights=None) # –í–µ—Å–∞ –Ω–µ –≥—Ä—É–∑–∏–º, –∑–∞–≥—Ä—É–∑–∏–º —Å–≤–æ–∏
model.classifier[1] = nn.Linear(model.classifier[1].in_features, 2)

# 3. –ó–∞–≥—Ä—É–∂–∞–µ–º –≤–µ—Å–∞
# map_location='cpu' –≤–∞–∂–Ω–æ, –µ—Å–ª–∏ –∫–æ–Ω–≤–µ—Ä—Ç–∏—Ä—É–µ–º –Ω–∞ –º–∞—à–∏–Ω–µ –±–µ–∑ GPU –∏–ª–∏ –ø–æ—Å–ª–µ –æ–±—É—á–µ–Ω–∏—è
model.load_state_dict(torch.load(MODEL_PATH, map_location='cpu'))
model.eval() # –ü–µ—Ä–µ–≤–æ–¥–∏–º –≤ —Ä–µ–∂–∏–º –∏–Ω—Ñ–µ—Ä–µ–Ω—Å–∞ (–≤—ã–∫–ª—é—á–∞–µ–º Dropout –∏ —Ç.–¥.)

# 4. –°–æ–∑–¥–∞–µ–º —Ñ–µ–π–∫–æ–≤—ã–π –≤—Ö–æ–¥–Ω–æ–π —Ç–µ–Ω–∑–æ—Ä (Dummy Input)
# –†–∞–∑–º–µ—Ä: (Batch_Size=1, Channels=3, Height=224, Width=224)
dummy_input = torch.randn(1, 3, IMG_SIZE, IMG_SIZE)

# 5. –≠–∫—Å–ø–æ—Ä—Ç –≤ ONNX
torch.onnx.export(
    model,                  # –ú–æ–¥–µ–ª—å
    dummy_input,            # –ü—Ä–∏–º–µ—Ä –≤—Ö–æ–¥–∞
    ONNX_PATH,              # –ö—É–¥–∞ —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å
    export_params=True,     # –°–æ—Ö—Ä–∞–Ω–∏—Ç—å –≤–µ—Å–∞ –≤–Ω—É—Ç—Ä–∏ —Ñ–∞–π–ª–∞
    opset_version=12,       # –í–µ—Ä—Å–∏—è ONNX
    do_constant_folding=True,
    input_names=['input'],  # –ò–º—è –≤—Ö–æ–¥–∞ (–≤–∞–∂–Ω–æ –¥–ª—è –∏–Ω—Ñ–µ—Ä–µ–Ω—Å–∞!)
    output_names=['output'], # –ò–º—è –≤—ã—Ö–æ–¥–∞
    dynamic_axes={'input': {0: 'batch_size'}, 'output': {0: 'batch_size'}} # –ß—Ç–æ–±—ã –º–æ–∂–Ω–æ –±—ã–ª–æ –∫–∏–¥–∞—Ç—å –±–∞—Ç—á–∏ —Ä–∞–∑–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞
)

print(f"‚úÖ Model converted to {ONNX_PATH}")
```

```python
import onnxruntime as ort
import numpy as np
from PIL import Image
import os

# –ù–∞—Å—Ç—Ä–æ–π–∫–∏
ONNX_PATH = "leukemia_model.onnx"

def preprocess(img_path):
    img = Image.open(img_path).convert('RGB')
    img = img.resize((224, 224))
    
    # Preprocessing –∫–∞–∫ –ø—Ä–∏ –æ–±—É—á–µ–Ω–∏–∏ (Normalize ImageNet)
    x = np.array(img, dtype=np.float32) / 255.0
    mean = np.array([0.485, 0.456, 0.406], dtype=np.float32)
    std = np.array([0.229, 0.224, 0.225], dtype=np.float32)
    x = (x - mean) / std
    
    # Transpose (H, W, C) -> (C, H, W)
    x = x.transpose(2, 0, 1)
    
    # Add batch dimension -> (1, C, H, W)
    x = np.expand_dims(x, axis=0)
    return x.astype(np.float32)

# –ò—â–µ–º —Ç–µ—Å—Ç–æ–≤—É—é –∫–∞—Ä—Ç–∏–Ω–∫—É
img_path = None
for root, dirs, files in os.walk("./local_dataset_cache_optimized"):
    for file in files:
        if file.endswith(".jpg"):
            img_path = os.path.join(root, file)
            break
    if img_path: break

if img_path:
    print(f"Testing on image: {img_path}")
    
    # –ó–∞–ø—É—Å–∫ ONNX Runtime
    session = ort.InferenceSession(ONNX_PATH)
    input_name = session.get_inputs()[0].name
    
    X = preprocess(img_path)
    outputs = session.run(None, {input_name: X})
    logits = outputs[0][0]
    
    # Softmax
    probs = np.exp(logits) / np.sum(np.exp(logits))
    
    classes = ['Normal (HEM)', 'Leukemia (ALL)']
    prediction = classes[np.argmax(probs)]
    
    print(f"\nüß† Prediction: {prediction}")
    print(f"üìä Probabilities: {classes[0]}={probs[0]:.4f}, {classes[1]}={probs[1]:.4f}")
else:
    print("‚ùå No images found for testing.")
```
